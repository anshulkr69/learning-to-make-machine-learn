# **IMPORTS**

```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

# **Actual Scraping**

```python
name=[]
add=[]
city=[]
pin=[]
state=[]
cp=[]
purp=[]
aim=[]
web=[]
tel=[]
mob=[]

for j in range(1,13):
  temp_name=[]

  response=requests.get('https://ngosindia.org/category/jharkhand-ngos/ranchi-ngos/page/{}'.format(j))

  soup=BeautifulSoup(response.text,'lxml')
  temp_name=soup.find_all('h3',class_='loop-title')

  for i in range(len(temp_name)):
    name.append(temp_name[i].text.strip())
```

```python
for s in name:
  response=requests.get('https://ngosindia.org/jharkhand-ngos/{}'.format(s.replace(' ','-').replace(',','')[:-7]))
  soup=BeautifulSoup(response.text,'lxml')

  raw_det=soup.find_all('p')[1].get_text(separator='\n',strip=True)
  detail=raw_det.split('\n')
  add.append(detail[0][5:] if detail[0][5:] else None)
  try:
    city.append(detail[1] if detail[1] else None)
  except:
    city.append(None)
  try:
    pin.append(detail[2][5:] if detail[2][5:] else None)
  except:
    pin.append(None)
  try:
    state.append(detail[3] if detail[3] else None)
  except:
    state.append(None)
  try:
    cp.append(detail[4][16:] if detail[4][16:] else None)
  except:
    cp.append(None)
  try:
    purp.append(detail[5][9:] if detail[5][9:] else None)
  except:
    purp.append(None)
  try:
    aim.append(detail[6][23:] if detail[6][23:] else None)
  except:
    aim.append(None)
  try:
    web.append(detail[7][9:] if detail[7][9:] else None)
  except:
    web.append(None)
  try:
    tel.append(detail[8][5:] if detail[8][5:] else None)
  except:
    tel.append(None)
  try:
    mob.append(detail[9][8:].strip() if detail[9][9:] else None)
  except:
    mob.append(None)
```

```python
NGO_Ranchi=pd.DataFrame({'Name':name,'Address':add,'City':city,'Pin':pin,'State':state,'Contact Person':cp,'Purpose':purp,'Aim/Objective/Mission':aim,'Website URL':web,'Telephone No.':tel,'Mobile No.':mob})
```
