## **MAE** (Mean Absolute Error)

$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$

Advantage:


*  Same unit as data
*   Robust to outliers

Disadvantage:


*   The graph is not differentiable






## **MSE** (Mean Squared Error)

$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$


Advantage:

*   Differentiable function

Disadvantage:

*   Not robust to outliers






## **RMSE** (Root Mean Squared Error)

$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$


Advantage:

*   Differential function
*   Same unit as data

Disadvantage:

*   More robust than MSE still not as robust to outliers



## **R2 Score** (How good is our model performing)

- it also called coeff. of determinants

- or goodness of fit

- basically tells how good the fit/regression line is as compared to mean line.

### $R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$

y_cap is for predicted line

y_bar is for mean line


Special Cases:


*   when score is zero: Its not utilizing the parameters
*   when score is one: Its overfitted/Over utilizing the parameter

**That means the more we move towards 1 the better our model is!!!**




Another way to look at this score is:
lets say there is a dataset having CGPA, IQ and LPA, and the R2 Score is 0.80. This means that the model can explain 80% of why the LPA is that much or 80% of the result depends on CGPA and IQ but it cannot explain the rest 20%. It may be from external factors/uncontrollable factors.
