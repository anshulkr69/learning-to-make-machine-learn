## Feature selection
Not every feature/column in a dataset are useful. These useless features just make the model inefficient, so its better to identify the useful features and train our model on those features.

RandomForest helps us identify the important features by giving us the feature importance score.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/mnist_train_small.csv', header = None)
df.shape
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs = 5)
```

```python
rf.fit(X, y)
```

```python
rf.feature_importances_
```

```python
rf.feature_importances_.shape
```

```python
feat_sel = np.reshape(rf.feature_importances_,(28,28))
```

```python
pd.DataFrame(feat_sel)
```

```python
sns.heatmap(feat_sel)
```

So this heatmap shows us the columns of importance i.e. the darker the pixel, the less important it is.

## How these algo determine feature importance?

For this we need to see how DecisionTree finds the importance of a feature.

## $\Delta I_{\text{norm}}(t) = \frac{N_t}{N_{\text{total}}} I(t) - \frac{N_{t_L}}{N_{\text{total}}} I(t_L) - \frac{N_{t_R}}{N_{\text{total}}} I(t_R)
$


$\Delta I_{\text{norm}}(t)$: The normalized node importance for node t.

$N_{\text{total}}$: The total number of samples in the entire dataset.

$N_t, N_{t_L}, N_{t_R}$: The number of samples at node t, its left child, and its right child.

$I(t), I(t_L), I(t_R)$: The impurity values of the parent node and its respective children.


Using this formula the importance is calculated for each splitting node and then noted to which column does it belong i.e. on which column did that node split on. When every calculation is complete, then we move on to calculate the feature importance by summing up all the node importance that belongs to specific feature/column and is divided by the total sum of all node importance.

A similar thing happens in RandomForest, the column-wise importance is calculated for ever column and their mean is used to evaluate the final feature importance.
