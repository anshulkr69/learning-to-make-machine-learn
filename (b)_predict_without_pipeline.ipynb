```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('/content/drive/MyDrive/Models/ohe_sex.pkl','rb'))
ohe_emb = pickle.load(open('/content/drive/MyDrive/Models/ohe_emb.pkl','rb'))
clf = pickle.load(open('/content/drive/MyDrive/Models/cfl.pkl','rb'))
```

```python
# Assume user input
# Pclass/gender/age/SibSp/Parch/Fare/Embarked
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1,7)
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
test_input_sex
```

```python
test_input_embarked = ohe_emb.transform(test_input[:,-1].reshape(1,1))
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
# very important to maintain the order of array in which we performed the actual fitting and transformation
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```
