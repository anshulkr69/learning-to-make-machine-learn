## **Univariate Data Analysis**

This is performed on two types of Data:
1. Categorical/Discrete
2. Numerical/Continuous

### **For categorical data**

```python
# COUNTPLOT
sns.countplot(df['total_rooms'])   #NOTE: there is no categorical data in our dataset but this will give
# a good idea on how this works

#since there is a lot of variation in the total rooms column therefor the plot is pretty dense
# not of much help for data analysis
```

```python
# Another way to COUNTPLOT
df['total_rooms'].value_counts().plot(kind='pie',autopct='%.2f')
# kind can also be 'bar' or many other things, look up the plot method for more info
```

### For Numerical Data

```python
# HISTOGRAM
plt.hist(df['population'])
```

```python
sns.distplot(df['population'])

# This is a probability distribution function curve.
```

The `df.skew()` function in Pandas calculates the skewness of numerical data in a DataFrame. Skewness measures the asymmetry of a probability distribution.

### **Interpretation of Skewness Values:**


*   Skewness = 0: Perfectly symmetrical distribution (e.g., standard
normal distribution).
*   Skewness > 0: Positively (right) skewed. The right tail is longer, meaning the majority of data points cluster on the left.
*   Skewness < 0: Negatively (left) skewed. The left tail is longer, meaning the majority of data points cluster on the right.










```python
df['population'].skew()
```

# **Bivariate Data Analysis**

analyzing data of two columns is **`bivariate analysis`** and more than that is multivariate.

When we are dealing with two columns we can get various relationships between data:

1.   Numerical-Numerical
2.   Numerical-Categorical
3.   Catergorical-Categorical



```python
tips=sns.load_dataset('tips')
tips.sample(5)
```

```python
flights=sns.load_dataset('flights')
flights.sample(5)
```

```python
iris=sns.load_dataset('iris')
iris.sample(5)
```

## 1. Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(tips,x=tips['total_bill'],y=tips['tip'])
# this is bivariate, lets do multivariate
sns.scatterplot(tips,x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
# hue show difference using color, style uses marker, size uses marker size
```

## 2. Bar Plot (N-C)

```python
sns.barplot(tips,x=tips.day,y=tips.total_bill,hue=tips.time)
```

## 3. Box plot (N-C)

```python
sns.boxplot(flights,x=flights.month,y=flights.passengers,)
# Hue can also be added here
```

## 4. Distplot (N-C)

```python
titanic=pd.read_csv('train.csv')
```

```python
titanic.sample(5)
```

```python
sns.kdeplot(titanic[titanic.Survived==0].Age,)
sns.kdeplot(titanic[titanic.Survived==1].Age,)
```

there is a clear trend that the younger children have higher chance of survival.
Old aged people have higher chance of not surviving.

## Heatmaps (categorical-categorical)

```python
# first we need  to create a crossmap of the datas that we want on a heatmap.
pd.crosstab(titanic.Pclass,titanic.Survived)
```

```python
sns.heatmap(pd.crosstab(titanic.Pclass,titanic.Survived))
```

heatmaps on look good with big databases having good amount of categorical data like the flight database

```python
tab_flight=flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(tab_flight)
```

```python
# The percentage way
titanic.replace(['female','male'],[1,0]).groupby('Embarked')[['Sex','Survived']].mean()*100
# we did this replacing thing because the mean will act like mean and if we want to know
# how many female passengers are boarded from various ports we will need to conver females
# to 1 and all of the other genders to 0
```

```python
titanic.groupby('Embarked')['Pclass'].value_counts()
```

### clustermaps (C-C)

```python
pd.crosstab(titanic.Parch,titanic.Survived)
sns.clustermap(pd.crosstab(titanic.Parch,titanic.Survived))
```
